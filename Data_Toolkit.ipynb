  Q2)  How does broadcasting work in NumPy?

  Ans: How broadcasting works:

1. Matching dimensions: When operating on two arrays, NumPy compares their shapes element-wise from right to left (i.e., from the last dimension to the first).
2. Size 1 dimensions: If one array has a dimension of size 1, it can be broadcasted to match the corresponding dimension of the other array.
3. Singleton dimensions: If one array has fewer dimensions than the other, NumPy adds singleton dimensions (of size 1) to the left of the array's shape until the number of dimensions matches.

Q6)  What are the differences between NumPy arrays and Python lists?

Ans: NumPy arrays and Python lists are two different data structures in Python that serve distinct purposes.

Differences:

1. Purpose:
    - NumPy arrays: Designed for numerical computations and scientific computing.
    - Python lists: General-purpose data structures for storing and manipulating collections of data.
2. Data type:
    - NumPy arrays: Homogeneous, meaning all elements must be of the same data type (e.g., integers, floats).
    - Python lists: Heterogeneous, meaning elements can be of different data types (e.g., integers, strings, objects).
3. Memory usage:
    - NumPy arrays: More memory-efficient, especially for large datasets, due to their homogeneous nature and compact storage.
    - Python lists: Less memory-efficient, as each element is stored as a separate object with additional metadata.
4. Performance:
    - NumPy arrays: Optimized for numerical computations, providing faster performance for operations like matrix multiplication and element-wise operations.
    - Python lists: Slower performance for numerical computations, but suitable for general-purpose programming and data manipulation.
5. Operations:
    - NumPy arrays: Support vectorized operations, allowing element-wise operations on entire arrays.
    - Python lists: Require explicit loops or list comprehensions for element-wise operations.

Q21)  Why is NumPy’s array slicing faster than Python’s list slicing?

Ans: NumPy's array slicing is faster than Python's list slicing for several reasons:

1. Memory layout: NumPy arrays store data in a contiguous block of memory, which allows for faster access and manipulation of data. In contrast, Python lists store data in a non-contiguous manner, which can lead to slower performance.
2. Vectorized operations: NumPy arrays support vectorized operations, which allow operations to be performed on entire arrays at once. This can lead to significant performance improvements compared to Python's list slicing, which requires iterating over individual elements.
3. Native code: NumPy's array slicing is implemented in C, which is a lower-level language that can execute operations more efficiently than Python. In contrast, Python's list slicing is implemented in Python itself, which can lead to slower performance.
4. Less overhead: NumPy's array slicing has less overhead than Python's list slicing, as it doesn't require creating new Python objects or performing type checking.

Q4)  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame
np.random.seed(0)
data = np.random.rand(100, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

Q12)  Use Pandas to load a CSV file and display its first 5 rows?

```python
import pandas as pd

# Load the CSV file
def load_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print("The file does not exist.")
        return None
    except pd.errors.EmptyDataError:
        print("The file is empty.")
        return None
    except pd.errors.ParserError:
        print("Error parsing the file. Please check the file format.")
        return None

# Display the first 5 rows
def display_first_rows(df):
    if df is not None:
        print(df.head(5))

# Usage
file_path = 'data.csv'  # replace with your CSV file path
df = load_csv(file_path)
display_first_rows(df)
```

Q13)  Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate some random 3D data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y,
    'z': z
})

# Create the 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z')

# Update the layout
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
